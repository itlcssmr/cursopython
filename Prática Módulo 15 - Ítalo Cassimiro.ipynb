B) Já é possível identificar variáveis com possíveis outliers? Se sim, quais?

Sim, o maior valor *total_pago* de 8564.75 indica a possibildiade de existência de outliers quando consideramos o valor médio pago de 2290.09.

```python
# B) Já é possível identificar variáveis com possíveis outliers? Se sim, quais?

# Variável tempo_como_cliente ----------------------------------------------------------------------------

# Calculando os quartis e IQR para a variável 'total_pago'
Q1 = df['total_pago'].quantile(0.25)
Q3 = df['total_pago'].quantile(0.75)
IQR = Q3 - Q1

# Determinando os limites inferior e superior
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Identificando os outliers
outliers = df[(df['total_pago'] < limite_inferior) | (df['total_pago'] > limite_superior)]

# Exibindo os outliers encontrados
print('Variável tempo_como_cliente ---------------------------')
outliers
```

```python
# Variável pagamento_mensal ----------------------------------------------------------------------------

# Calculando os quartis e IQR para a variável 'pagamento_mensal'
Q1 = df['pagamento_mensal'].quantile(0.25)
Q3 = df['pagamento_mensal'].quantile(0.75)
IQR = Q3 - Q1

# Determinando os limites inferior e superior
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Identificando os outliers
outliers = df[(df['pagamento_mensal'] < limite_inferior) | (df['pagamento_mensal'] > limite_superior)]

# Exibindo os outliers encontrados
print('Variável pagamento_mensal ---------------------------')
outliers
```

```python
# Variável total_pago ----------------------------------------------------------------------------

# Calculando os quartis e IQR para a variável 'total_pago'
Q1 = df['total_pago'].quantile(0.25)
Q3 = df['total_pago'].quantile(0.75)
IQR = Q3 - Q1

# Determinando os limites inferior e superior
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Identificando os outliers
outliers = df[(df['total_pago'] < limite_inferior) | (df['total_pago'] > limite_superior)]

# Exibindo os outliers encontrados
print('Variável total_pago ---------------------------')
outliers
```
